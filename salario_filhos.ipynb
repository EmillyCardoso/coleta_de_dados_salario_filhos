```python
def coleta_dados(): #Irá sair duas listas
  sal_mens = []
  num_filhos = []
  salario = 0.0
  filhos = -1
  while salario >= 0.0:# o exercicio pede que se faça isso até o salario estar negativo
     salario = float(input("Digite o valor do salário(-1 para terminar): "))
     if salario >=0:
       sal_mens.append(salario)
       while filhos < 0:
         filhos = int(input("Digite o número de filhos: "))
       num_filhos.append(filhos)
       filhos = -1
  return (sal_mens, num_filhos)

def calcula_media(lista):
  soma = 0
  for elemento in lista:
    soma += elemento
  return soma/len(lista)

def calc_perc(num_filhos, sal_mensais, qtde_filhos=0, salario_menor=0):
  qtde_pessoas = 0
  for i in range(len(num_filhos)):
    if num_filhos[i]>qtde_filhos and sal_mensais[i] < salario_menor:
      qtde_pessoas += 1
  perc_pessoas = (qtde_pessoas*100)/len(num_filhos)
  return perc_pessoas

def calcula_maior_lista(lista): #de forma recursiva
  if len(list) == 1:
    return lista[0]
  else:
    if lista[0] > calcula_maior_lista(lista[1:]):
      return lista[0]
    else:
      return calcula_maior_lista(lista[1:])

def calcula_maior_lista2(lista):
  maior = lista[0]
  for n in lista:
    if n > maior:
      maior = n
  return maior

def cal_perc_sal(sal_mensais, salario_menor=0):
  cont = 0
  for sal in sal_mensais:
    if sal <= salario_menor:
      cont += 1
  perc_menor = (cont*100)/len(sal_mensais)
  return perc_menor



salarios_mensais, numeros_filhos = coleta_dados()
media_salario_populacao = calcula_media(salarios_mensais)
media_num_filhos        = calcula_media(numeros_filhos)
percen_pop_maisde2filhos_menosde1500 = calc_perc(numeros_filhos, salarios_mensais, qtde_filhos=2, salario_menor=1500)
maior_salario = calcula_maior_lista2(salarios_mensais)
perc_sal_800 = cal_perc_sal(salarios_mensais, salario_menor=800)

print(salarios_mensais, numeros_filhos)
print(media_salario_populacao, media_num_filhos)
print(percen_pop_maisde2filhos_menosde1500)
print("Maior salario = ", maior_salario)
print("A média de salario da população éigual a R$",media_salario_populacao)
print("A média de numero de filhos é igual a ", media_num_filhos, " por habitante")
print("Pessoas com mais de 2 filhos e que recebem menos de R$ 1500,00 corresponde a ", percen_pop_maisde2filhos_menosde1500, "% da amostragem")
print(perc_sal_800, "% da população sobrevivem com até R$ 800,00 por mês")
```
